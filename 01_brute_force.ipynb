# Brute Force

総当たりで探索する手法
テキストを並べて、1文字ずつ頭からパターンが合致するか確認する
- メモリ: 
- 計算量: O( |Text| * |Patterns| )
- デメリット: 大量のデータに対して計算量が膨大になる


引用: https://www.coursera.org/learn/algorithms-on-strings/lecture/K12kF/brute-force-approach-to-pattern-matching

```python
def brute_force(text, pattern):
    index = []
    for i, str_t in enumerate(text):
        for j, str_p in enumerate(pattern):
            if text[i+j] != str_p:
                break
            if j == len(pattern) - 1:
                index.append(i)
    return index
```

## brownコーパスでテスト

```python
with open('./sample.txt', 'r') as f:
    text = f.read()
pattern = 'do'
```

```python
brute_force(text, pattern)
```

```python
patterns = [
    'do',
    'what',
    'make'
]
%time
for pattern in patterns:
    print(brute_force(text, pattern))
```
